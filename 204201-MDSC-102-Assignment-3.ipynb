```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Binomial distribution
f = lambda n, p, x: np.math.comb(n, x) * p**x * (1-p)**(n-x)

def binomial(n, p):
    result = [f(n, p, x) for x in range(n+1)]

    plt.bar(range(n+1), result)
    plt.xlabel('successes')
    plt.ylabel('Probability')
    plt.title('Binomial Distribution')
    plt.grid(True)
       
binomial(30, 0.2)
```

```python
#Poisson distribution
f = lambda lam, x: (np.exp(-lam) * lam**x) / np.math.factorial(x)

def poisson(lam):
    result = [f(lam, x) for x in range(0, 20)]
        
    plt.bar(range(0, 20), result)    
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.title('Poisson Distribution')
    plt.grid(True)

poisson(5)
```

```python
#Exponential distribution
f = lambda lam, x: (lam*np.exp(-lam*x))

def exponential(lam):
    result = [f(lam, x) for x in range(1, 20)]
    
    plt.plot(range(1, 20), result)    
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.title('Exponential Distribution')
    plt.grid(True)

exponential(1)
```

```python
#Normal distribution
f = lambda mean, variance, x: (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-0.5 * ((x - mean) / np.sqrt(variance))**2)

def normal(mean, variance):
    x_values = np.linspace(-10, 10, 1000) 
    result = [f(mean, variance, x) for x in x_values]
       
    plt.plot(x_values, result)
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.title('Normal Distribution')
    plt.grid(True)

normal(1, 2)
```

```python
#Gamma function
f = lambda alpha, lamb, x: (lamb**alpha * x**(alpha-1) * np.exp(-lamb * x)) / np.math.gamma(alpha)

def gamma(alpha, lamb):
    x_values = np.linspace(0, 10, 1000)  
    result = [f(alpha, lamb, x) for x in x_values]
        
    plt.plot(x_values, result)
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.title('Gamma Distribution')
    plt.grid(True)
    
gamma(2, 1)
```
