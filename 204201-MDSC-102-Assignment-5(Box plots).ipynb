```python
attributes =('class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanine', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline')
df.columns = attributes
df  #assigning names/attributes to all columns
```

```python
# plotting count plot 
sns.countplot(data=df, x='class')
```

```python
#plotting scatter plot with alcohol as x axis and ash as y axis
sns.scatterplot(data = df, x = 'Alcohol', y = 'Ash') 
plt.title('Scatter plot for Alcohol and Ash')
plt.show() 
```

```python
plt.subplots #plotting scatter plot with alcohol as x axis and ash as y axis along with displaying class and changing colors of points
sns.scatterplot(data=df, x = 'Alcohol', y = 'Ash', hue='class', palette=['green','black','red'])
```

```python
plt.subplots
sns.scatterplot(data=df, x = 'Alcohol', y = 'Malic acid', hue='class', palette=['green','black','red'])
```

```python
plt.subplots
sns.scatterplot(data=df, x = 'Alcohol', y = 'Alcalinity of ash', hue = 'class', palette=['green','black','red'])
```

```python
fig, axes = plt.subplots(4,3, figsize=(15,15)) #number of rows and columns along with mentioning size 
fig.subplots_adjust(hspace=0.5, wspace=0.5) #adjust distances between each plot(height and width)
i=j=0 
for feature in df.columns:
    if j>2:
        j=0
        i+=1
        
    if feature != 'Alcohol' and  feature != 'class':
        sns.scatterplot(data=df, x = 'Alcohol', y = feature, hue= 'class', ax = axes[i,j], palette = ['black', 'red', 'green'])
        axes[i,j].set_title('Alcohol and {} '. format(feature))
        j+=1
```

Inferences made from the above scatterplots 

1. Class 1 has more alcohol content, while class 2 has comparitively more content of malic acid, and class 3 is widely spread.
2. Alcohol content is high in class 1, while class 2 has comparitively more content of Ash.
3. Alcalinity of Ash content is low in class 1, while class 2 has comparitively more content of Alcalinity of Ash.
4. Magnesium content is exceptionally high compared to alcohol.
5. Alcohol content as well as total phenols content is very high in class 1.
6. Flavanoids content is high in class 2.
7. Alcalinity of Ash content is low in class 1, while class 2 has comparitively more content of Alcalinity of Ash.
8. Proline content is exceptionally high when compared to alcohol.


 Overall , Class 1 has more alcohol of all the plots.

```python
fig , axes = plt.subplots(12,3,figsize=(30,60))
fig.subplots_adjust(hspace=0.5,wspace=0.5)

def hist(df,i):

  j = 0

  for col in df.columns:

    if col != 'class' and col != 'Alcohol':
      sns.histplot(ax=axes[i,j],kde=True,data=df,x=col)
      j = j + 1

    if j > 2:
      j = 0
      i = i + 1

hist(df[df['class']==1],0)
hist(df[df['class']==2],4)
hist(df[df['class']==3],8)
```
